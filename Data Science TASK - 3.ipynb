# Importing Libraries

```python
import numpy as np
import cv2
from PIL import Image
from IPython.display import display
```

```python
image=cv2.imread("img.png")
image=cv2.resize(image,(200,300))
rgb_image=cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
display(Image.fromarray(rgb_image))
```

# Converting Image to Greyscale Image

```python
greyscale_img=cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
display(Image.fromarray(greyscale_img))
```

# Converting the Image to Negative Image

```python
neg_img=cv2.bitwise_not(greyscale_img)
display(Image.fromarray(neg_img))
```

# Smoothening the image

```python
smooth_img=cv2.GaussianBlur(neg_img, (5,5),0,0)
display(Image.fromarray(smooth_img))
```

# Converting Image into Pencil Sketch

```python
sketch_img=255 - smooth_img
sketch=cv2.divide(greyscale_img, sketch_img, scale=250.0)
display(Image.fromarray(sketch))
```
